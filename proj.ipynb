```python
# 1 Passo instal as lib (pacotes) => pip install pandas(trab com BD) numpy openpyxl nbformat ipykernel plotly(trab com grafico)
# 2 Passo de selacionar o PY que quer usar => acima do V 3.10
# 3 Passo de importar a BD (pandas) e cria a variavel, ver tabela, criar uma variavel


import pandas as pd # para carregar a BD

#pandas ler e armazenar minha BD, o pd se conecta com arq fora do seu PC
tabela = pd.read_csv("cancelamentos.csv") # essa variavel tabela vai guardar a BD

# colunas inuteis (imp tirar essas colunas ex id do cliente ela nao ajuda em nada...)colunas inuteis que atrapalham
#deletar colunas
tabela = tabela.drop(columns="CustomerID") # minha tabela velha eh a nova tabela, remover a coluna CustomerID

#ver a tabela
#print(tabela)
#usar o comando display, ele ajuda a ver melhor a table
display(tabela)



```

```python
# Passo 5: Corrigir problemas da base de dados
# valores vazios na tabela
# colunas inuteis (imp tirar essas colunas ex id do cliente ela nao ajuda em nada...)colunas inuteis que atrapalham
#deletar colunas
# tabela = tabela.drop(columns="CustomerID") # minha tabela velha eh a nova tabela, remover a coluna CustomerID

# valores vazios na tabela
display(tabela.info()) #mostra todos os valores e todos preenchidos na tabela com os dados
#float = numero com casa decimal
# object = coluna com valores de texto

#corrig valores vazios, como fazer isso? escolher como vc vai fazer isso de acordo com a situacao
#como nao interfere a informacao apos analise posso excluir as linhas, pois nao eh um informcao import para analise
#deletar linhas com valore vazios
tabela = tabela.dropna() #NaN => Not a Number = valores vazios, apaga todas linhas NaN



```

```python
# Passo 6: Análise dos cancelamentos = qtos clients cancelaram e qual a % de clientes (0 nao cance e 1 canc) analise na coluna
#contar os valores da coluna cancelou (selecionar uma coluna da tabela )
#mostra os valores da coluna cancelou
display(tabela["cancelou"].value_counts()) # coluna cancelou da tabela passa entre conchetes, no PY  toda lista eh entre colchetes
#1.0    14950 CANCELOU - VERDADEIRO cancelou  -mas a empresa vai informar
#0.0    11269  NAO  CANCELOU - FALSO - mas a empresa vai informar

#em percentual
display(tabela["cancelou"].value_counts(normalize=True)) #calculando a proporcao de um valor dentro da coluna
#se passar o false ele nao calcula em Porcentagem
```

```python
#contar qtos clentes tem na colua assintura
display(tabela["assinatura"].value_counts())
```

```python
#saber a media como cliente
display(tabela["tempo_como_cliente"].mean())
```
